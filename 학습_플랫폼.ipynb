```python
import os
import random
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)

def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)

seed_everything(42)

import warnings
warnings.filterwarnings('ignore')
```

```python
train_org = pd.read_csv('train.csv')
test_org = pd.read_csv('test.csv')

display(train_org.head())
display(test_org.head())
```

```python
train_df = train_org.copy()
test_df = test_org.copy()
```

```python
test_x = test_df.drop(columns=['user_id']).copy()
train_x = train_df[test_x.columns].copy()
train_y = train_df['target'].copy()
```

```python
!pip install category_encoders

from sklearn.preprocessing import LabelEncoder
from category_encoders.target_encoder import TargetEncoder

categorical_features = list(train_x.dtypes[train_x.dtypes == "object"].index)

display(categorical_features)

for i in categorical_features:
    le = TargetEncoder(cols=[i])
    train_x[i] = le.fit_transform(train_x[i], train_y)
    test_x[i] = le.transform(test_x[i])

display(train_x.head())
display(test_x.head())
```

```python
train_x.fillna(0, inplace=True)
test_x.fillna(0, inplace=True)
```

```python
import tensorflow as tf

def rmsle(y_true, y_pred):
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(y_pred, tf.float32)

    squared_error = tf.square(tf.math.log1p(y_pred) - tf.math.log1p(y_true))

    return tf.sqrt(tf.reduce_mean(squared_error))

def loss_fn(y_true, y_pred):
    return rmsle(y_true, y_pred)

def metric_fn(y_true, y_pred):
    return rmsle(y_true, y_pred)
```

```python
callbacks_list = [
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=30, verbose=2, mode='min',restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.8, patience=3, min_lr=0.00001),
    tf.keras.callbacks.TerminateOnNaN()
]
```

```python
import tensorflow as tf
import pandas as pd

# Assume that 'train_x', 'train_y', 'callbacks_list', 'loss_fn', and 'metric_fn' are defined.

# Learning Rate 및 Epoch 수정
def create_model():
    input_layer = tf.keras.Input(shape=(len(train_x.columns),))
    x = tf.keras.layers.BatchNormalization(epsilon=0.00001)(input_layer)
    x = tf.keras.layers.Dense(16, activation='relu')(x)
    x = tf.keras.layers.Dense(64, activation='relu')(x)  # 수정된 부분: 뉴런 수를 64로 증가시킴
    x = tf.keras.layers.Dropout(0.5)(x)  # 수정된 부분: 50%의 드롭아웃 추가
    x = tf.keras.layers.Dense(32, activation='relu')(x)
    output_layer = tf.keras.layers.Dense(1)(x)

    model = tf.keras.Model(inputs=input_layer, outputs=output_layer)

    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  loss=loss_fn,
                  metrics=[metric_fn]
                  )

    return model

# 더 많은 에폭으로 학습
model = create_model()
history = model.fit(train_x.astype('float32'), train_y.astype('float32'),
                    epochs=300,  # 수정된 에폭 수
                    callbacks=callbacks_list,
                    validation_split=0.1)
```

```python
# 예측 및 submission4.csv로 저장
sample_submission = pd.read_csv('sample_submission.csv')
predictions = model.predict(test_x.astype('float32'))
binary_predictions = (predictions >= 0.5).astype(int)
sample_submission["target"] = binary_predictions
sample_submission.to_csv("submission4.csv", index=False)

# 파일을 Colab 다운로드 디렉토리로 이동
!mv submission2.csv /content/

# 파일 다운로드 링크 제공
from google.colab import files
files.download("/content/submission2.csv")
```
